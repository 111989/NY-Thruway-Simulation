```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import statsmodels.api as sm 
import numpy as np
from scipy import stats
import warnings
from scipy.stats import chisquare

file = pd.read_csv("final_study_data.csv")
file_copy = file.copy()

vehicle_class = file["vehicle_class"].unique()
vehicle_class.sort()

entrance_site = file["entrance_site"].unique()
entrance_site.sort()
```

```python
entrance = []
vclass = []
lamda = []
intercept = []
alpha = []
chisqstat = []
pvalue = []

for i in range(len(entrance_site)):
    for j in range(len(vehicle_class)):
        file = file_copy[(file_copy["entrance_site"] == entrance_site[i]) & (file_copy["vehicle_class"] == vehicle_class[j]) & (file_copy["year"] == 2017) & (file_copy["payment_type"] == "E-ZPass")]
        x = file["week_of_year"].unique()
        x.sort()
        dataframe = file.drop(["exit_site","payment_type","profit","year"], axis = 1).groupby("week_of_year").mean()
        y = dataframe["vehicle_count"].values
        
        model_p = smf.poisson("vehicle_count ~ 1", data = dataframe)
        result_p = model_p.fit()
        lamda_ = float(np.exp(result_p.params))
        
        entrance.append(entrance_site[i])
        vclass.append(vehicle_class[j])
        lamda.append(lamda_)
        
        minimum = int(y.min())
        maximum = int(y.max())
        while minimum % 10 != 0:
            minimum -= 1
        while maximum % 10 != 0:
            maximum += 1
        X = stats.poisson(lamda_)
        v, k = np.histogram(y, range = (minimum, maximum), density = True)
        chisquare_ = chisquare(v, f_exp = X.pmf(k[:-1]))[0]
        pvalue_ = chisquare(v, f_exp = X.pmf(k[:-1]))[1]
        chisqstat.append(chisquare_)
        pvalue.append(pvalue_)
        
        model_nb = smf.negativebinomial("vehicle_count ~ 1", data = dataframe, loglike_method='nb2')
        result_nb = model_nb.fit()
        intercept.append(float(result_nb.params.values[0]))
        alpha.append(float(result_nb.params.values[1]))
        
dataframe17 = {
    "entrance_site": entrance,
    "vehicle_class": vclass,
    "lambda": lamda,
    "intercept": intercept,
    "alpha": alpha,
    "chisquare": chisqstat,
    "pvalue": pvalue
}
dataframe17 = pd.DataFrame(dataframe17, columns = ["entrance_site", "vehicle_class", "lambda", "intercept", "alpha", "chisquare", "pvalue"])
dataframe17.to_csv(r"C:\Users\16315\Desktop\Projects\AMS 553\dataframe17.txt", sep = "\t")
```
